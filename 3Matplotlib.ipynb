# Matplotlib


Matplotlib is the default plotting library for Python.

In this notebook, we'll discuss some basics of the Matplotlib Library.


````python
import numpy as np
import matplotlib.pyplot as plt       # This is the usual shorthand notation

# This is an IPython magic command. You can also try `%matplotlib notebook`
%matplotlib inline
````

## Line Plot

Here, we present an example to quickly plot lines.

````python
# get x and y coordinates of a few points
x = np.arange(0, 5) 
y = np.sin(x)

# draw lines connecting the points 
plt.plot(x, y)

# display the plot
plt.show()
````

PyPlot provides a MATLAB-like plotting framework. In the above example, we selected a few points that lie on a sinusoidal function. Plot function draws lines connecting each point sequentially. Plot function also takes other arguments, such as, color, linestyle, etc. Here is an example that uses a different set of arguments for the plot function:

````python
# draw lines connecting the points 
plt.plot(x, y,  color='r', linestyle='--', marker='o')

# display the plot
plt.show()
````

The above code displays a dashed red plot with circular markers at each point. You can experiment changing these to style your plots differently.

## Scatter Plot 

Here, we'll discuss how to plot points without drawing lines connecting them.

````python
# get random coordinates for 50 points
N = 50
x = np.random.rand(N)
y = np.random.rand(N)

# get random colors to use for these points
colors = np.random.rand(N)

# get random areas of each point (displayed as a shaded circle). Area is given by pi*r^2
area = np.pi * (15 * np.random.rand(N))**2  # 0 to 15 point radiuses

# plot all the points scattered across the figure
plt.scatter(x, y, s=area, c=colors, marker='o')

# disply the plot
plt.show()
````

Scatter function takes similar arguments like the plot function. You can change the marker to other types, e.g. '*', 'D', etc. 

## Image Plot

Here, we'll discuss how to draw or display image plots.

````python
import matplotlib.cbook as cbook

# get a sample image to display
image_file = cbook.get_sample_data('logo2.png')

# load the image from file
image = plt.imread(image_file)

# plot the image
plt.imshow(image)

# clear x- and y-axes
plt.axis('off')  

# display the plot
plt.show()
````

## Bar Plots

You can make bar plots to display statistical data. Here is an example:

````python
# setup some statistical data
N = 5
group1_means = (20, 35, 30, 35, 27)
group1_std = (2, 3, 4, 1, 2)

group2_means = (25, 32, 34, 20, 25)
group2_std = (3, 5, 2, 3, 3)

# the x locations for the bars
ind = np.arange(N)  

# the width of the bars
width = 0.35       

# plot first group
plt.bar(ind, group1_means, width, color='r', yerr=group1_std)

# plot second group
plt.bar(ind + width, group2_means, width, color='y', yerr=group2_std)

# add some text for labels, title and axes ticks
plt.ylabel('Scores')
plt.title('Scores by group and types')
plt.xticks(ind + width, ('T1', 'T2', 'T3', 'T4', 'T5'))

# display plot legend
plt.legend(('Group1', 'Group2'))

# display the plot
plt.show()
````

The above example displays a bar plot comparing some statistical data collected for two groups. Bar function's arguments are: (locations, height, width, color, errorlines, ...). The errorlines is an optional argument used to display standard deviation lines. 

## Subplots

Here, we will present an example to setup multiple subplots within a single plot figure.

````python
# data for the plots
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

# subplot 1
plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

# subplot 2
plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

# display the plots
plt.show()
````

The arguments of the subplot function used here are as follows: (num of rows, num of columns, plot number). Plot number starts with 1 and has to be less than or equal to the product of the number of rows and columns. 

There is a huge list of different plot styles in matplotlib. For a comprehensive set of examples check out: http://matplotlib.org/gallery.html

## Using style sheets
Sometimes, we don't want fine control over our plot, but there are some plotting settings that we want to apply to all of our plots (e.g., a new color cycle, changing the axes spines to dark grey, etc.). The easiest way to do that is to use style sheets. Matplotlib has several built in.

To use a style sheet for all plots in a session:
```python
plt.style.use(<sheet_name>)
```
If you want to use a style sheet on just a single plot without changing the default plot settings you can use a context manager:
```python
with plt.style.context(<sheet_name>):
    plt.plot(...)
```

**Exercise:** Use a context manager with the codes above using one of the built-in styles. You can find available built-in styles using the `plt.style.available` command. 


````python
# show all styles
print(plt.style.available)

# set global default style
plt.style.use("fivethirtyeight")

# plot a single plot with a diffrent style
with plt.style.context('dark_background'):
    plt.plot(x1,y1, label='Test')
    plt.legend()
    plt.show()

# plot another plot with the default style
plt.plot(x2,y2)
plt.show()
````

HINT: this can take a long time to compute, so frist implement it for a smaller number of trials and then once you are happy with the plots you can increase it.

````python
# taken from the first Notebook
class Vector:
    def __init__(self, coords=[0,0,0]):
        self.coords = coords
        
    def norm(self):
        sum = 0
        for x in self.coords:
            sum += x**2
        return sum**(1/2)
    
    def scalar_multiplication(self, value):
        return Vector([value*x for x in self.coords])
    
    def inner_product(self, vec):
        if len(vec.coords) != len(self.coords):
            raise ValueError
        sum = 0
        for i, x in enumerate(self.coords):
            sum += x * vec.coords[i]
        return sum    
````

````python
import time
n = 50
ind = np.arange(1, 20+1)

my_means = []
my_std = []
np_means = []
np_std = []

for i in ind:
    my_times = []
    np_times = []
    for j in range(n):
        # generate random Vecotors of dimension 2**i
        arr1 = np.random.randint(0, 10, 2**i)
        arr2 = np.random.randint(0, 10, 2**i)
        v1 = Vector(arr1)
        v2 = Vector(arr2)

        # test my implementation
        t = time.time_ns()
        v1.inner_product(v2)
        my_times.append(time.time_ns() - t)

        # test numpy's implementation
        t = time.time_ns()
        np.inner(arr1, arr2)
        np_times.append(time.time_ns() - t)

    # calculate and remember means and standard deviation for i
    my_means.append(np.mean(my_times))
    my_std.append(np.std(my_times))
    np_means.append(np.mean(np_times))
    np_std.append(np.std(np_times))
````

4. Sample $100$ pairs of vectors of increasing size $d=\{2, 4, 8, 16, 32, 64, 128, 256, 512, 1024\}$ from a multivariate normal distribution with zero mean identity covariance (`np.identity()`). For each value of $d$, plot a histogram with 20 bins of the pairwise angles between the pairs of vectors sampled. You should end up with 10 histograms, each displayed in a subplot.

````python
d = [2**i for i in np.arange(1, 10+1)]
N = 100
n_bins = 20


for (i, n) in enumerate(d):
    v1 = np.random.multivariate_normal([0 for i in range(n)], np.identity(n), N)
    v2 = np.random.multivariate_normal([0 for i in range(n)], np.identity(n), N)

    # make unit vectors instead of dividing by the multiplication of the vector's length
    v1 /= np.linalg.norm(v1, axis=1).reshape(N, 1)
    v2 /= np.linalg.norm(v2, axis=1).reshape(N, 1)

    angles = np.rad2deg(np.arccos(np.sum(v1 * v2, axis=1)))
    
    # plot the distribution of angles
    plt.subplot(3, 4, i+1)
    plt.title(f'd={n}')
    plt.xticks(np.round(np.linspace(np.min(angles)+0.5, np.max(angles)-0.5, 3))) # +-0.5 to ensure they are visible on the plot
    plt.hist(angles, bins=n_bins)

plt.tight_layout()
plt.show()
````
